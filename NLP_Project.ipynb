## Import Libraries

```python
import os
import string
import re
import nltk
from nltk.tokenize import word_tokenize # for splitting a large sample of text into words
from nltk.stem.isri import ISRIStemmer  # Stemming Lib
from pyarabic.araby import * #A specific Arabic language library for Python
```

## Load Data

```python
def load_data():
    
    path = "./Arabic"     # Dataset PATH
    
    data = ""
    
    for file in os.listdir(path):      #Loop on files as list
        
        if file.endswith('.txt'):      # Chech if File End With TXT
            
            file_path = os.path.join(path,file)   # add PATH with file 
            
            with open(file_path , encoding="utf-8-sig") as f: # Read Text inside Files ,UTF-8: to support Arabic Language 
                
                for lines in f:  # Loop in Lines inside the File
                    
                    lines = lines.strip() # Separate Lines using strip()
                    
                    data += lines      # Add Lines to data , data Variable: store all lines of all dataset
    return data
```

```python
arabic_pun = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
english_pun = string.punctuation
pun_list = arabic_pun + english_pun
```

## Preprocessing

```python
def preprocessing(data):
    
    # Convert Or Replace pun_list with Empty Value with the same size of pun_list
    translator = str.maketrans(pun_list , ' ' * len(pun_list)) 
    
    data = data.translate(translator)
    
    # Repalce Numbers, English Char with Empty Value
    data = re.sub('[0-9]+',' ',data)
    data = re.sub('[A-Z]+',' ',data)
    data = re.sub('[a-z]+',' ',data)

    return data
```

```python
ara_txt = load_data()
ara_txt = preprocessing(ara_txt)
```

## Normaliztion

```python
def normalize_arabic(text):
    
    # حذف الزخرفات المختلفة على النص 
    text = strip_diacritics(text)
    
    # حذف التطويل أو الشدة
    text = strip_tatweel(text)
    
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("ك", "گ", text)
    
    return text
```

```python
txt = normalize_arabic(ara_txt)
word_tokenize(txt)
```

## Stemming

```python
def stemming():
    st = ISRIStemmer()
    w = txt
    for a in word_tokenize(w):
        print(st.stem(a))
```
